## Section1

```python
project = ("Space Recommendation", "1.0", 2026, "python", "Yerim")

print("=" * 50)
print(f"   {project[0]} - Open Source Project")
print(f"   Project Lead : {project[-1]}")
print("=" * 50)

print("Each field by index")
print(project[0], project[1], project[2], project[3], project[4])

print(f"Name : {project[0]}  |  Version : {project[1]}  |  Started : {project[2]}")
print(f"First 3 fields  : {project[:3]}")
print(f"Language count  : {project.count('python')}    Language index : {project.index('python')}")
print("# project[1] = 2.0 -> TypeError: 'tuple' object does not support item assignment -> Because tuple is immutable.")
print("-" * 50)

contributors = []

for _ in range(4):
    name = input("Name: ")
    role = input("Role: ")
    language = input("Language: ")
    commits = int(input("Commits: "))
    country = input("Country: ")
    contributor = {"name" : name,
                   "role" : role,
                   "language" : language,
                   "commits" : commits,
                   "country" : country
                   }
    contributors.append(contributor)

names = [c["name"] for c in contributors]
names.sort()
print(f"Sorted names    : {names}")
print(f"Last name       : {names[-1]}           First two : {names[:2]}")

for c in contributors:
    c.update({"status" : "Active"})
print(f"Contributor 1 status : {contributors[0].get('status')}")

cont1 = contributors[0].copy()
print(f"Backup : {cont1}")
```

## Section2

```python
issues = []
for _ in range(5):
    id = int(input("Id: "))
    title = input("Title: ")

    type = input("Type (Bug/Feature): ")
    while type.lower() not in ['bug', 'feature']:
        type = input("Invalid input. Please type 'Bug/'Feature': ")

    priority = input("Priority (Critical/High/Medium/Low): ")
    while priority.lower() not in ['critical', 'high', 'medium', 'low']:
        priority = input("Invalid input. Please type 'Critical/High/Medium/Low': ")
    
    reporter = input("Reporter: ")
    status = input("Status (Open/In Progress/Resolved): ")
    while status.lower() not in ['open', 'in progress', 'resolved']:
        status = input("Invalid input. Please type 'Open/In Progress/Resolved': ")
    issue = {"id" : id,
             "title" : title,
             "type" : type,
             "priority" : priority,
             "reporter" : reporter,
             "status" : status
            }
    issues.append(issue)

open_count = 0
for issue in issues:
    if issue.get('status') == 'Open':
        open_count += 1
print(f"Open issues  : {open_count}")
issues[0]["priority"] = "Critical"
print("First issue -> priority updated to Critical.")
print(f"Last two issues : {issues[-2:]}")


print("-" * 50)
reporters = set()
for issue in issues:
    reporters.add(issue['reporter'])
print(f"reporters   : {reporters}")

tech_stack = set()
for c in contributors:
    tech_stack.add(c['language'])
print(f"tech_stack  : {tech_stack}")
tech_stack.add("CSS")
tech_stack.discard("HTML")

reporter_tech = reporters.union(tech_stack)
print(f"union       : {reporter_tech}")
print(f"intersection: {reporters.intersection(tech_stack)}   difference : {reporters.difference(tech_stack)}")
print(f"Critical present  : {'Yes' if (any(issue['priority'] == 'Critical' for issue in issues)) else 'No'} - flag for immediate review.")

print("-" * 50)
priority_count = {}
for issue in issues:
    p = issue['priority']
    if p not in priority_count:
        priority_count[p] = 1
    else:
        priority_count[p] += 1

print(f"priority keys : {priority_count.keys()}")
print(f"priority values : {priority_count.values()}")


status_open = []
status_in_progress = []
status_resolved = []
for issue in issues:
    if issue['status'] == "Open":
        status_open.append(issue['title'])
    elif issue['status'] == "In Progress":
        status_in_progress.append(issue['title'])
    else:
        status_resolved.append(issue['title'])
        
status_groups = {}
if len(status_open)!=0:
    status_groups['Open'] = status_open
if len(status_in_progress)!=0:
    status_groups['In Progress'] = status_in_progress
if len(status_resolved)!=0:
    status_groups['Resolved'] = status_resolved
print(f"Status_Groups -> {'   |   '.join(f'{k} : {v}' for k, v in status_groups.items())}")
print(f"Status Keys : {status_groups.keys()}")
print(f"Status Values : {status_groups.values()}")

top_count = 0
for reporter in reporters:
    cnt = 0
    for issue in issues:
        if issue['reporter'] == reporter:
            cnt += 1
    if cnt > top_count: 
        top_count = cnt
        top_reporter = reporter
print(f"Top reporter   : {top_reporter} ({top_count} issues)")

issues[0].pop("type")
print(f"After pop('type') : {issues[0]}")
```

## Section3

```python
import os
#print(f"Current directory: {os.getcwd()}")

## create folder_name
folder_name = project[0].lower().replace(" ", "_")

## create folder
if not os.path.exists(folder_name):
    os.mkdir(folder_name)
    print(f"Folder created: {folder_name}/")
else:
    print(f"Folder '{folder_name}'  already exists")

## create file path
report_path = os.path.join(folder_name, 'project_report.txt')
issues_path = os.path.join(folder_name, 'issues.csv')
print(report_path)
print(priority_count)

## create file with wraping try/except IOError
try:
    with open(report_path, "w") as f:
        f.write("=" * 50 + "\n")
        f.write("Space - Recommendation - Open Source Project\n")
        f.write("Project Lead : Yerim\n")
        f.write("=" * 50 + "\n")

        f.write("\n======= Project Info =======\n")   
        f.write(f"Project : {project[0]}   Version : {project[1]}   Lead : {project[4]}\n")

        f.write("\n======= Contributors =======\n")
        f.write(f"Contributors : {len(contributors)}   Names : {[d['name'] for d in contributors]}\n")
        for c in contributors:
            f.write(f"{c}\n")

        f.write("\n======= Issues =======\n")
        f.write(f"Issues : {len(issues)}  Open: {sum(1 for issue in issues if issue['status']=='Open')}  Reporters : {len(reporters)}\n")
        for issue in issues:
            f.write(f"{issue}\n")

        f.write("\n======= Analysis =======\n")
        f.write(f"Tech Stack : {tech_stack}\n")
        f.write(f"{'  '.join(f'{k}: {v}' for k, v in priority_count.items())}\n")

        f.write("\n======= Reporters =======\n")
        f.write(f"Reporters : {reporters}\n")
        f.write(f"Top Reporter : {top_reporter} ({top_count} issues)\n")

except IOError:
    print("Error in 'project_report.txt'")

try:
    with open(issues_path, "w") as f: 
        f.write("id, title, priority, reporter, status\n")
        for issue in issues:
            f.write(f"{', '.join(f'{v}' for v in issue.values())}\n")
except IOError:
    print("Error in 'issues.csv'")


print(f"Files saved : {os.listdir(folder_name)}")


## read file with wraping try/except FileNotFoundError
for file in os.listdir(folder_name):
    file_path = os.path.join(folder_name, file)
    try:
        with open(file_path, "r") as f:
            print("\n\n" + "-" * 40)
            print("File:", file)

            print("---------------- read() ---------------- ")
            print(f.read())

        with open(file_path, "r") as f:
            print("-------------- readline() --------------")
            print(f"Line 1 : {f.readline()}")
            print(f"Line 2 : {f.readline()}")

        with open(file_path, "r") as f:
            print("-------------- readlines() --------------")
            lines = f.readlines()
            print(f"Total lines : {len(lines)}   Critical/High lines : {len([l for l in lines if 'Critical' in l or 'High' in l])}")
            print(f"Critical/High lines : {[l for l in lines if 'Critical' in l or 'High' in l]}")
    except FileNotFoundError:
        print(f"File '{file}' Not Found")



## Final
print("="  * 50)
print(f"             {project[0]}")
print("="  * 50)
print(f"Project : {project[0]}     Version : {project[1]}     Lead : {project[4]}")
print(f"Contributors : {len(contributors)}     Names : {names}")
print(f"Issues : {len(issues)}     Open : {sum(1 for d in issues if d['status'] == 'Open')}     Reporters : {len(reporters)}")
print(f"Tech Stack : {tech_stack}")
print(f"Reporters : {reporters}")
print(f"Top Reporters : {top_reporter} ({top_count} issues)")
print(f"{'   '.join(f'{k}: {v}' for k, v in priority_count.items())}")
print(f"Report : {report_path}")
print(f"CSV : {issues_path}")
print("="  * 50)
print(f"{project[0]} complete. Thank you for contributing to open source!")
```
